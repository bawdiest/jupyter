## Load and Prepare Weather Data

```python
e_weatherLog2016 = pd.read_csv("plant.Watering.data/shopdwhdata_0YH_WAE.csv", 
                           sep = ";", 
                           header = 0, 
                           skip_blank_lines=True) #2016 Data
e_weatherLog2014 = (pd.read_csv("plant.Watering.data/shopdwhdata_2YH_WAE.csv", 
                                sep = ";", header = 0, skip_blank_lines=True))

e_weatherLog = e_weatherLog2014
e_weatherLog = e_weatherLog.append(e_weatherLog2016, ignore_index = True)

list(e_weatherLog)
print(e_weatherLog.dtypes)
```

```python
e_weatherLog["DateTime"] = pd.to_datetime(e_weatherLog.date + ' '  + e_weatherLog.time)
e_weatherLog = e_weatherLog.set_index(e_weatherLog.DateTime)

e_weatherLog
```

```python
# Select relevant variables.
p_weatherLog = pd.DataFrame(e_weatherLog)
p_weatherLog = p_weatherLog.loc[:,["DateTime", "tre200b0","ure200b0","rre150b0","sre000b0","fu3010b0","prestab0"]]

p_weatherLog = p_weatherLog.dropna(axis = 0, how = "any")

# Set "right" varialbe type
p_weatherLog.sre000b0 = pd.to_numeric(p_weatherLog.sre000b0, errors='coerce')
p_weatherLog.fu3010b0 = pd.to_numeric(p_weatherLog.fu3010b0, errors='coerce')
#p_weatherLog.dkl010b0 = pd.to_numeric(p_weatherLog.dkl010b0, errors= 'coerce')
p_weatherLog.fu3010b0 = pd.to_numeric(p_weatherLog.fu3010b0, errors= 'coerce')
#p_weatherLog.fu3010b1 = pd.to_numeric(p_weatherLog.fu3010b1, errors= 'coerce')

```
